```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.rand(100),
    'C': np.random.choice(['Group 1', 'Group 2', 'Group 3'], 100)
})
```

```python
data.head()
```

```python
sns.histplot(data['A'], kde = True)
plt.title('Distribution Plot')
plt.show()
```

```python
plt.scatter(data['A'], data['B'])
plt.title('Scatter Plot')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```python
sns.boxplot(x = 'C', y = 'A', data = data)
plt.title('Box Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['A'])
plt.title('Line with customizations')
plt.xlabel('Index')
plt.ylabel('Values')
plt.xlim(0, 100)
plt.ylim(data['A'].min(), data['A'].max())
plt.show()
```

```python
sns.scatterplot(x = 'A', y = 'B', hue = 'C', data = data)
plt.title('Scatter Plot with Hue')
plt.show()
```

```python
# Grouped charts
# Distribution of 2 continuous columns with respect to one categorical variable
sns.histplot(data[data['C'] == 'Group 1']['A'], label='Group 1',kde = True)
sns.histplot(data[data['C'] == 'Group 2']['A'], label='Group 2', kde = True)
sns.histplot(data[data['C'] == 'Group 3']['A'], label='Group 3', kde = True)
plt.legend()
plt.title('Grouped Distribution Plots')
plt.show()
```
